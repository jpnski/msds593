### 2) Decluttering

```python
months = pd.Series(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])

sales_18 = pd.Series([38, 47, 49, 42, 42, 38, 43, 52, 49, 67, 74, 86])
sales_19 = pd.Series([101, 78, 91, 77, 78, 71, 78, 62, 68, 50, 43, 41])
```

```python
fig, ax = plt.subplots()

ax.plot(months, sales_18, label='2018')
ax.plot(months, sales_19, label='2019')

ax.set_ylabel('Quantity Sold (Thousands)')
ax.grid(alpha=0.25)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.title('Vehicles Sold Per Month in 2018 and 2019')
ax.legend(loc=1)
plt.show()
```
